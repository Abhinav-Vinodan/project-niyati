```python
# Step 1: Business Understanding
# This dataset appears to be related to a marketing campaign, likely from a bank, aiming to predict customer response to term deposit offers.
# The target variable is 'y' (yes/no), indicating if a customer subscribed to a term deposit.

# Step 2: Domain Knowledge
# Common financial attributes: age, job, marital status, education, loan, housing, balance.
# Campaign-specific attributes: contact type, campaign duration, number of contacts, previous campaign outcomes.

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Data Cleaning

# Load the dataset (assuming the file is named 'test.csv')
file_path = 'test.csv'
data = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the dataset
print("Dataset Preview:")
print(data.head())

# Check data types and for missing values
print("\nData Info:")
print(data.info())

# Summary statistics for numerical columns
print("\nSummary Statistics for Numerical Columns:")
print(data.describe())

# Step 4: Data Analysis

# Target variable analysis
print("\nTarget Variable Distribution (y):")
y_distribution = data['y'].value_counts(normalize=True)
print(y_distribution)

# Categorical variable analysis
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome']
print("\nCategorical Variable Distributions:")
for col in categorical_cols:
    print(f"\n{col} distribution:")
    print(data[col].value_counts(normalize=True))

# Summary statistics for numerical columns
numerical_cols = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
print("\nNumerical Variable Summary:")
print(data[numerical_cols].describe())

# Step 5: Visualization (Optional)

# Visualization for the target variable
plt.figure(figsize=(6,4))
sns.countplot(x='y', data=data)
plt.title('Target Variable (y) Distribution')
plt.show()

# Visualization of categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='job', hue='y', data=data)
plt.title('Job Distribution by Target (y)')
plt.xticks(rotation=45)
plt.show()

# Histogram for numerical variables
plt.figure(figsize=(10, 6))
data[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features')
plt.show()

```
